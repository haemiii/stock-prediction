```python
from marcap import marcap_data
df = marcap_data('2020-10-01', '2020-10-10')
df
```

```python
import pandas as pd

year = [y for y in range(2009, 2020)]
month = [m for m in range(1, 13)]
cap = pd.DataFrame()

for i in year:
  for j in month:
    start_date = f'{i}-{j:02d}-01'  # 월을 두 자리로 포맷팅
    end_date = f'{i}-{j:02d}-10'

    df = marcap_data(start_date, end_date).head(10)
    df = df[['Name', 'Marcap']]
    cap = pd.concat([cap, df])
```
