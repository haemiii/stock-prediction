```python
import yfinance as yf
import pmdarima as pm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from pmdarima.arima import auto_arima
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 1.  ( 10)
stocks = ['', '', 'POSCO', '', '', 'SK', '', 'LG', '', 'NAVER']
ticker = ['005930', '005380', '005490', '012330', '015760', '000660', '055550', '051915', '032830', '035420']
st = {}

for i, t in enumerate(ticker):
    symbol = t + '.KS'
    data = yf.download(symbol, '2009-01-01', '2019-12-31')
    st[stocks[i]] = data

# 2. Adj Close  
plt.figure(figsize=(12, 8))
for k, v in st.items():
    v['Adj Close'].plot(label=k)
plt.legend()
plt.title('Stock Prices')
plt.show()
```

```python

# 3.  
for k, v in st.items():
    df_train, df_test = train_test_split(v, test_size=0.1, shuffle=False, random_state=42)

    # 4. ACF  ,  
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    plt.suptitle(k)
    plot_acf(df_train['Adj Close'], lags=40, ax=ax[0])
    plot_pacf(df_train['Adj Close'], lags=40, ax=ax[1])

    # 5.  
    diff_train = df_train['Adj Close'].diff().dropna()

    #   acf, pacf  
    fig, ax = plt.subplots(3, 1, figsize=(8, 10))
    ax[0].plot(diff_train.index, diff_train)
    plot_acf(diff_train, lags=40, ax=ax[1])
    plot_pacf(diff_train, lags=40, ax=ax[2])

    # 6. parameter search(   )
    model = auto_arima(
        df_train['Adj Close'],
        start_p=1, start_q=1, max_p=3, max_q=3,
        seasonal=False, trace=True,
        error_action='ignore', suppress_warnings=True, alpha=0.06
    )

    # 7.  
    forecast = model.predict(len(df_test), return_conf_int=True, start=df_test.index[0], end=df_test.index[-1])

    # 8.    
    forecast_value = forecast[0]
    forecast_lb = forecast[1][:, 0]
    forecast_ub = forecast[1][:, 1]
    forecast_index = list(df_test.index)

    fig, ax = plt.subplots(figsize=(12, 8))
    v['Adj Close'].plot()
    ax.vlines(forecast_index[0], 0, max(v['Adj Close']), linestyle='--', color='r', label='Start of Forecast')
    ax.plot(forecast_index, forecast_value, label='Prediction')
    ax.fill_between(forecast_index, forecast_lb, forecast_ub, color='k', alpha=0.1, label='0.95 Prediction Interval')
    ax.legend(loc='upper left')
    plt.suptitle(f'{k}\nARIMA {model.order}, Prediction Results')
    plt.show()

    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_squared_log_error
```

### RMSE

```python
import numpy as np
def performance_metrics(actual_values, forecast_values):
    rmse = np.sqrt(mean_squared_error(actual_values, forecast_values))
    mae = mean_absolute_error(actual_values, forecast_values)
    r_squared = r2_score(actual_values, forecast_values)
    msle = mean_squared_log_error(actual_values, forecast_values)

    return rmse, mae, r_squared, msle

#        
for k, v in st.items():
    df_train, df_test = train_test_split(v, test_size=0.1, shuffle=False, random_state=42)

    # ARIMA  
    model = auto_arima(
        df_train['Adj Close'],
        start_p=1, start_q=1, max_p=3, max_q=3,
        seasonal=False, trace=True,
        error_action='ignore', suppress_warnings=True
    )

    #  
    forecast = model.predict(len(df_test), return_conf_int=True, start=df_test.index[0], end=df_test.index[-1])
    forecast_values = forecast[0]
    actual_values = df_test['Adj Close'].values

    #   
    rmse, mae, r_squared, msle = performance_metrics(actual_values, forecast_values)

    #  
    print(f"Performance Metrics for {k}:")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"R-squared: {r_squared:.4f}")
    print(f"MSLE: {msle:.4f}")
    print()


```
